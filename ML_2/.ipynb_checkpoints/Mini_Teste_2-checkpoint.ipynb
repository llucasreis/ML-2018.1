```python
import pandas as pd
import numpy as np

from sklearn.metrics import f1_score, confusion_matrix
from sklearn.model_selection import GridSearchCV, KFold, train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, BaggingClassifier

import pydot
from sklearn import tree
from sklearn.externals.six import StringIO  

import seaborn as sns
from matplotlib import pyplot as plt
```

```python
file_path_red = 'data\winequality-red.csv'
file_path_white = 'data\winequality-white.csv'

red_df = pd.read_csv(file_path_red, sep=';')
white_df = pd.read_csv(file_path_white, sep=';')
```

# Criação do DataFrame

```python
df = pd.concat([red_df, white_df]) 
df.head()
```

# Histograma do atributo alvo

```python
plt.hist(df['quality'], color='DarkRed', ec='Black', bins=np.arange(0, 11, 1));

plt.xticks(np.arange(0, 11, 1));
plt.yticks(np.arange(0, 3500, 500));

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.xlabel('Qualidade', fontsize=17);
plt.ylabel('Quantidade de Observações', fontsize=17);
fig.savefig('img\\Histograma.png', dpi=250)
plt.title('Histograma do Atributo Qualidade', fontsize=18);

plt.show()
```

# Divisão dos atributos e definição de parâmetros 

```python
X_data, Y_data = df.drop(['quality'], axis=1), df['quality']

k_fold = KFold(n_splits=3, shuffle=True, random_state=42)
train_indexes, test_indexes = [], []

for train_index, test_index in k_fold.split(X_data):
    train_indexes.append(train_index)
    test_indexes.append(test_index)
```

```python
X_train, X_test = X_data.drop(test_indexes[1]), X_data.drop(train_indexes[1])
Y_train, Y_test = Y_data.drop(test_indexes[1]), Y_data.drop(train_indexes[1])
```

```python
values = list(x for x in range(1, len(df.columns)))
lr = np.arange(0.1, 1.1, 0.1)
class_values = list(x for x in range(0, 11))

tree_forest_params = {'criterion': ['gini', 'entropy'], 'max_depth': values, 'max_features': values}
adaboosting_params = {'algorithm': ['SAMME', 'SAMME.R'], 'learning_rate': lr}
bagging_params = {'bootstrap': ['True', 'False'], 'bootstrap_features': ['True', 'False'], 'max_features': values}
```

# Árvores de Decisão

```python
k_fold = KFold(n_splits=3, shuffle=True, random_state=42)
splits = k_fold.split(X_data)

GS_tree = GridSearchCV(DecisionTreeClassifier(), tree_forest_params, 'f1_micro', cv=splits, return_train_score=True)
GS_tree.fit(X_data, Y_data);
```

# Resultados

```python
tree_metrics = pd.DataFrame(GS_tree.cv_results_).drop(['params'], axis=1)
drop_columns, values = [], ['train']

for column in tree_metrics.columns:
    if (any(s in str(column) for s in values)): drop_columns.append(column)

tree_metrics.drop(drop_columns, axis=1, inplace=True)
tree_metrics.sort_values('rank_test_score', inplace=True)
tree_metrics.drop(['rank_test_score'], axis=1, inplace=True)

tree_metrics.head()
```

# Floresta Aleatória

```python
k_fold = KFold(n_splits=3, shuffle=True, random_state=42)
splits = k_fold.split(X_data)

GS_random_forest = GridSearchCV(DecisionTreeClassifier(), tree_forest_params, 'f1_micro', cv=splits, return_train_score=True)
GS_random_forest.fit(X_data, Y_data);
```

# Resultados

```python
random_forest_metrics = pd.DataFrame(GS_random_forest.cv_results_).drop(['params'], axis=1)
drop_columns, values = [], ['train']

for column in random_forest_metrics.columns:
    if (any(s in str(column) for s in values)): drop_columns.append(column)

random_forest_metrics.drop(drop_columns, axis=1, inplace=True)
random_forest_metrics.sort_values('rank_test_score', inplace=True, ascending=True)
random_forest_metrics.drop(['rank_test_score'], axis=1, inplace=True)

random_forest_metrics.head()
```

# Melhor Modelo Floresta Aleatória

```python
random_forest_metrics.head(1)
```

# Adaboosting

```python
k_fold = KFold(n_splits=3, shuffle=True, random_state=42)
splits = k_fold.split(X_data)

GS_adaboosting = GridSearchCV(AdaBoostClassifier(), adaboosting_params, 'f1_micro', cv=splits, return_train_score=True)
GS_adaboosting.fit(X_data, Y_data);
```

# Resultados

```python
adaboosting_metrics = pd.DataFrame(GS_adaboosting.cv_results_).drop(['params'], axis=1)
drop_columns, values = [], ['train']

for column in adaboosting_metrics.columns:
    if (any(s in str(column) for s in values)): drop_columns.append(column)

adaboosting_metrics.drop(drop_columns, axis=1, inplace=True)
adaboosting_metrics.sort_values('rank_test_score', inplace=True, ascending=True)
adaboosting_metrics.drop(['rank_test_score'], axis=1, inplace=True)

adaboosting_metrics.head()
```

# Melhor Modelo Adabooting

```python
adaboosting_metrics.head(1)
```

# Bagging

```python
k_fold = KFold(n_splits=3, shuffle=True, random_state=42)
splits = k_fold.split(X_data)

GS_bagging = GridSearchCV(BaggingClassifier(), bagging_params, 'f1_micro', cv=splits, return_train_score=True)
GS_bagging.fit(X_data, Y_data);
```

# Resultados

```python
bagging_metrics = pd.DataFrame(GS_bagging.cv_results_).drop(['params'], axis=1)
drop_columns, values = [], ['train']

for column in bagging_metrics.columns:
    if (any(s in str(column) for s in values)): drop_columns.append(column)

bagging_metrics.drop(drop_columns, axis=1, inplace=True)
bagging_metrics.sort_values('rank_test_score', inplace=True, ascending=True)
bagging_metrics.drop(['rank_test_score'], axis=1, inplace=True)

bagging_metrics.head()
```

# Melhor Modelo Bagging

```python
bagging_metrics.head(1)
```

# Comparação dos Modelos

```python
import warnings
warnings.filterwarnings('ignore')

models = ['tree', 'forest', 'adaboosting', 'bagging']
models_dict = {'tree' : tree_metrics, 
               'forest' : random_forest_metrics, 
               'adaboosting' : adaboosting_metrics, 
               'bagging' : bagging_metrics}

results_df = pd.DataFrame(columns=['model'])

for model, idx in zip(models, range(0, 4)):
    results_df = results_df.append(models_dict[model].iloc[0])
    results_df.model.iloc[idx] = model

drop_columns, values = [], ['param', 'train']
for column in results_df.columns:
    if (any(s in str(column) for s in values)): drop_columns.append(column)
        
results_df.drop(drop_columns, axis=1, inplace=True)    
results_df.sort_values('mean_test_score', inplace=True, ascending=False)    

display(results_df)
```
